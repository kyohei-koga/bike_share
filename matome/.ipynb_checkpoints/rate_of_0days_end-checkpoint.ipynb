### スタート、到着が多いステーションの0になる日数を調べる AM
- start 63,62,65,64,82
- end 69,70,50

```python
import os
path = '/home/kyohei/bike_share/data_split_2015/'
files = []
for x in os.listdir(path):
    if 'San Francisco' in x:
        files.append(x)
```

```python
station_id = [int(x[33:35]) for x in files]
```

```python
statuses = []
for import_file_name in files:
    data = pd.read_csv(path+import_file_name)
    statuses.append(data)
```

```python
#日数
num_of_days = 366
num_of_dayofweek = 262
```

```python
#0のデータに限定
start_list_0 = []
for status in statuses:
    status = status[status.bikes_available==0]
    start_list_0.append(status)
    
end_list_0 = []
for status in statuses:
    status = status[status.docks_available==0]
    end_list_0.append(status)
```

```python
#日付データの変換
for status in start_list_0:
    status['time'] = status['time'].apply(lambda x:pd.to_datetime(x))
    status['Date'] = status['time'].apply(lambda x:x.date())
    status['hour'] = status['time'].apply(lambda x:x.hour)
    status['dayofweek'] = status['time'].apply(lambda x:x.dayofweek)

for status in end_list_0:
    status['time'] = status['time'].apply(lambda x:pd.to_datetime(x))
    status['Date'] = status['time'].apply(lambda x:x.date())
    status['hour'] = status['time'].apply(lambda x:x.hour)
    status['dayofweek'] = status['time'].apply(lambda x:x.dayofweek)
```

```python
#曜日の限定
start_dayofweek = []
for status in start_list_0:
    dayofweek = status[status.dayofweek<=4]
    start_dayofweek.append(dayofweek)

end_dayofweek = []
for status in end_list_0:
    dayofweek = status[status.dayofweek<=4]
    end_dayofweek.append(dayofweek)
```

```python
#時間帯と曜日の限定
AM_start_list = []
for status in start_dayofweek:
    AM = status[(status.hour>=7) & (status.hour<=9)]
    AM_start_list.append(AM)
    
AM_end_list = []
for status in end_dayofweek:
    AM = status[(status.hour>=7) & (status.hour<=9)]
    AM_end_list.append(AM)
```

```python
#時間帯と曜日の限定
PM_start_list = []
for status in start_dayofweek:
    PM = status[(status.hour>=16) & (status.hour<=19)]
    PM_start_list.append(PM)
    
PM_end_list = []
for status in end_dayofweek:
    PM = status[(status.hour>=17) & (status.hour<=20)]
    PM_end_list.append(PM)
```

```python
#AMで30分以上空車があった日数
AM_rate_start = []
for status in AM_start_list:
    rate = (status['Date'].value_counts()>=30).sum() / float(262)
    AM_rate_start.append(rate)
    
AM_rate_end = []
for status in AM_end_list:
    rate = (status['Date'].value_counts()>=30).sum() / float(262)
    AM_rate_end.append(rate)
```

```python
#PMで30分以上空車があった日数
PM_rate_start = []
for status in PM_start_list:
    rate = (status['Date'].value_counts()>=30).sum() / float(262)
    PM_rate_start.append(rate)
    
PM_rate_end = []
for status in PM_end_list:
    rate = (status['Date'].value_counts()>=30).sum() / float(262)
    PM_rate_end.append(rate)
```

```python
c = pd.DataFrame([station_id,AM_rate_start,AM_rate_end,PM_rate_start,PM_rate_end]).T
c.columns = ['station_id','AM_0_rate_bikes','AM_0_rate_docks','PM_0_rate_bikes','PM_0_rate_docks']
c['station_id'] = c['station_id'].astype(int)
c.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

c[c.station_id<90].sort_values(by='station_id',ascending=False).plot(x='station_id',y=['AM_0_rate_bikes','AM_0_rate_docks','PM_0_rate_bikes','PM_0_rate_docks'],kind='barh',figsize=(10,20))
plt.legend(loc='upper right')
```
