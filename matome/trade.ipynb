```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
station = pd.read_csv('/home/kyohei/bike_share_data/201608_station_data.csv')

station_SF = station[station.landmark=='San Francisco'][['station_id','landmark']]
```

```python
trip = pd.read_csv('/home/kyohei/bike_share_data/201608_trip_data.csv')

trip_SF = trip.merge(station_SF,left_on='Start Terminal',right_on='station_id',how='inner')
```

```python
trip_SF['Start Date'] = trip_SF['Start Date'].apply(lambda x:pd.to_datetime(x))

trip_SF['dayofweek'] = trip_SF['Start Date'].apply(lambda x:x.dayofweek)

trip_SF_dayofweek = trip_SF[trip_SF['dayofweek']<=4]
```

```python
trip_SF_dayofweek['start_hour'] = trip_SF_dayofweek['Start Date'].apply(lambda x:x.hour)
```

```python
trip_SF_dayofweek_AM = trip_SF_dayofweek[(trip_SF_dayofweek.start_hour>=7) & (trip_SF_dayofweek.start_hour<=9)]

start_end_group = pd.DataFrame(trip_SF_dayofweek_AM.groupby(['Start Terminal','End Terminal']).size()).reset_index()

start_end_group.rename(columns={0:'count'},inplace=True)

start_end_pivot = trip_SF_dayofweek_AM.pivot_table('Trip ID',index='Start Terminal',columns='End Terminal',aggfunc='count').fillna(0)
```
