```python
status_id50 = pd.read_csv('/home/kyohei/bike_share/data_split_2015/San Francisco_station_status_stid50.csv')

status_id50 = status_id50[['station_id','bikes_available','docks_available','time']]
```

```python
status_id50.info()
```

```python
#datetime
status_id50['time'] = status_id50['time'].apply(lambda x: pd.to_datetime(x))

#曜日
status_id50['dayofweek'] = status_id50['time'].apply(lambda x:x.dayofweek)

import datetime

#時間
status_id50['hour_time'] = status_id50['time'].apply(lambda x:datetime.time(x.time().hour,x.time().minute))

#日付
status_id50['Date'] = status_id50['time'].apply(lambda x:x.date())
```

```python
#稼働率の計算
status_id50['op_rate_bikes'] = status_id50.bikes_available / float(station[station.station_id==50].dockcount.values)
status_id50['op_rate_docks'] = status_id50.docks_available / float(station[station.station_id==50].dockcount.values)
```

```python
#曜日別、時間帯別、平均稼働率の計算
data_series_bikes = pd.DataFrame(status_id50.groupby(['dayofweek','hour_time'])['op_rate_bikes'].mean()).reset_index()
data_series_docks = pd.DataFrame(status_id50.groupby(['dayofweek','hour_time'])['op_rate_docks'].mean()).reset_index()
```

```python
fig, axes = plt.subplots(3,2,figsize=(15,10))
dayofweeks = [0,1,2,3,4]
targets = zip(dayofweeks, axes.flatten())
for (dayofweek,ax) in targets:
    data_series_bikes[data_series_bikes.dayofweek==dayofweek].plot('hour_time','op_rate_bikes',ax=ax)
    data_series_docks[data_series_docks.dayofweek==dayofweek].plot('hour_time','op_rate_docks',ax=ax)
    ax.set_ylim([0,1])
```

```python
#ある一週間
a_week_data = status_id50[(status_id50.Date>=datetime.date(2015,9,14)) & (status_id50.Date<=datetime.date(2015,9,18))]
```

```python
def plot_series(status_data,station,station_id,start_date=datetime.date(2015,9,14),end_date=datetime.date(2015,9,18)):
    #datetimeに変換
    status_data['time'] = status_data['time'].apply(lambda x: pd.to_datetime(x))
    
    #曜日情報の追加
    status_data['dayofweek'] = status_data['time'].apply(lambda x:x.dayofweek)
    
    #時間情報の追加
    status_data['hour_time'] = status_data['time'].apply(lambda x:datetime.time(x.time().hour,x.time().minute))
    
    #日付情報の追加
    status_data['Date'] = status_data['time'].apply(lambda x:x.date())
    
    #稼働率の計算
    status_data['op_rate_bikes'] = status_data.bikes_available / float(station[station.station_id==station_id].dockcount.values)
    status_data['op_rate_docks'] = status_data.docks_available / float(station[station.station_id==station_id].dockcount.values)
    
    #曜日別、時間帯別、平均稼働率の計算
    data_series_bikes = pd.DataFrame(status_data.groupby(['dayofweek','hour_time'])['op_rate_bikes'].mean()).reset_index()
    data_series_docks = pd.DataFrame(status_data.groupby(['dayofweek','hour_time'])['op_rate_docks'].mean()).reset_index()
    
    #曜日別、時間帯別、平均稼働率の推移図
    fig, axes = plt.subplots(3,2,figsize=(15,10))
    dayofweeks = [0,1,2,3,4]
    targets = zip(dayofweeks, axes.flatten())
    for (dayofweek,ax) in targets:
        data_series_bikes[data_series_bikes.dayofweek==dayofweek].plot('hour_time','op_rate_bikes',ax=ax)
        data_series_docks[data_series_docks.dayofweek==dayofweek].plot('hour_time','op_rate_docks',ax=ax)
        ax.set_ylim([0,1])
    
    #ある一週間
    a_week_data = status_data[(status_data.Date>=start_date) & (status_data.Date<=end_date)]
    
    #ある一週間の推移図
    fig, axes = plt.subplots(3,2,figsize=(15,10))
    dayofweeks = [0,1,2,3,4]
    targets = zip(dayofweeks, axes.flatten())
    for (dayofweek,ax) in targets:
        a_week_data[a_week_data.dayofweek==dayofweek].plot('hour_time','op_rate_bikes',ax=ax)
        a_week_data[a_week_data.dayofweek==dayofweek].plot('hour_time','op_rate_docks',ax=ax)
        ax.set_ylim([0,1])
```
