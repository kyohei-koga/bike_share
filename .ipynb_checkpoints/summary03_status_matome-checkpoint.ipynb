```python
status.time = status.time.apply(lambda x:pd.to_datetime(x))
```

```python
import datetime
#曜日
status['dayofweek'] = status.time.apply(lambda x:x.dayofweek)
#時間
status['hour_minute'] = status.time.apply(lambda x:datetime.time(x.time().hour,x.time().minute))
```

```python
station = pd.read_csv('/home/kyohei/bike_share_data/201608_station_data.csv')
```

```python
status['op_rate_bikes'] = status.bikes_available / station[station['station_id']==69]['dockcount'].values
status['op_rate_docks'] = status.docks_available / station[station['station_id']==69]['dockcount'].values
```

```python
status.head()
```

```python
#平均的振る舞い
acerage_behavior_bikes = pd.DataFrame(status.groupby(['dayofweek','hour_minute'])['op_rate_bikes'].mean()).reset_index()
acerage_behavior_docks = pd.DataFrame(status.groupby(['dayofweek','hour_minute'])['op_rate_docks'].mean()).reset_index()
```

```python
fig, axes = plt.subplots(3,2,figsize=(15,10))
dayofweeks = [0,1,2,3,4]
targets = zip(dayofweeks,axes.flatten())
for (dayofweek,ax) in targets:
    acerage_behavior_bikes[acerage_behavior_bikes.dayofweek==dayofweek].plot('hour_minute','op_rate_bikes',ax=ax)
    acerage_behavior_docks[acerage_behavior_docks.dayofweek==dayofweek].plot('hour_minute','op_rate_docks',ax=ax)
    ax.set_ylim([0,1])
```
