```python
hour_start = trip['Start Date'].apply(lambda x: x.hour)
hour_end = trip['End Date'].apply(lambda x: x.hour)
```

```python
#スタート時間と結合
trip_date_dayofweek_hour = pd.concat([trip_date_dayofweek,hour_start],axis=1)
#列名の変更
trip_date_dayofweek_hour.rename(columns={'Start Date':'hour'},inplace=True)
#曜日別、時間別の貸出台数
trip_grouby_dayofweek_hour = pd.DataFrame(trip_date_dayofweek_hour.groupby(['dayofweek','hour']).size()).reset_index()

fig, axes = plt.subplots(3,3,figsize=(15,8))
dayofweeks = [0,1,2,3,4,5,6]
target = zip(dayofweeks,axes.flatten())
for i,(dayofweek,ax) in enumerate(target):
    trip_grouby_dayofweek_hour[trip_grouby_dayofweek_hour.dayofweek==dayofweek].plot('hour',0,title=dayofweek,ax=ax)
    ax.set_xticks(range(25))
```

- 平日の利用ピークは8時と17時なので、多くは通勤での利用と考えられる
- 休日は11-16時の間が利用ピーク、遊びの移動手段として利用されていると考えられる。

```python
#スタート時間と結合
trip_date_dayofweek_hour = pd.concat([trip_date_dayofweek,hour_end],axis=1)
#列名の変更
trip_date_dayofweek_hour.rename(columns={'End Date':'hour'},inplace=True)
#曜日別、時間別の貸出台数
trip_grouby_dayofweek_hour = pd.DataFrame(trip_date_dayofweek_hour.groupby(['dayofweek','hour']).size()).reset_index()

fig, axes = plt.subplots(3,3,figsize=(15,8))
dayofweeks = [0,1,2,3,4,5,6]
target = zip(dayofweeks,axes.flatten())
for i,(dayofweek,ax) in enumerate(target):
    trip_grouby_dayofweek_hour[trip_grouby_dayofweek_hour.dayofweek==dayofweek].plot('hour',0,title=dayofweek,ax=ax)
    ax.set_xticks(range(25))
```

- ピークの位置は気持ちずれている感じがするが、ほとんど同じ。貸出から30分で終わるからであろう。
